```python
from bs4 import BeautifulSoup
```

```python
import requests
response = requests.get('https://www.ebay.com/')
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# Find all image tags
images = soup.find_all('img')
# Extract image sources
image_sources = [image.get('src') for image in images]
image_sources
```

```python
# Find all link tags
links = soup.find_all('a')

# Extract link hrefs
link_hrefs = [link.get('href') for link in links]
link_hrefs
```

```python
# Find all divs with the class 's-item__info'
divs = soup.find_all('div', {'class': 's-item__info'})

# Print the contents of the divs
for div in divs:
    print(div.text.strip())
```

```python
#extract specific class
div = soup.find('div', {'class': "vl-flyout-nav__expander"})
print(div.text.strip())
```

```python
import pandas as pd
data = []

# Loop through the links and extract the text and URL
for link in links:
    text = link.text.strip()
    href = link.get('href')
    data.append({'Text': text, 'URL': href})

```

```python
df = pd.DataFrame(data)
# Print the DataFrame
print(df)
```
